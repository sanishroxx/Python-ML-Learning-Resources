```python
#Step 1: Import necessary libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline 
from scipy.stats import pearsonr
import seaborn as sns  
```

```python
#Step 2: Reading the dataset 
df = pd.read_csv("loan-approval-dataset.csv")
```

```python
#Step 3: Understanding the dataset, finding shape of thedataset, info on the dataset, correlation among thevariables etc.
df.shape
```

```python
x = df.drop('loan-approval', axis=1)
y=df['loan-approval']
```

```python
print(x)
```

```python
print(y)
```

```python
#Create Train and Test set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1234)
```

```python
std_reg = LinearRegression()
```

```python
std_reg.fit(x_train,y_train)
```

```python
y_pred = std_reg.predict(x_test)

print(y_pred)
```

```python
mlr_score = std_reg.score(x_test,y_test)
print(mlr_score)
```

```python
plt.plot(month_number , total_profit , linestyle= '--' , color='red', linewidth = 3 , marker = 'o', mfc='k' ,label ="Profit data of last year",)
plt.xlabel("Month Number")
plt.ylabel("Sold units nuber")
plt.title("Company Sales Data of Last Year")
plt.legend( loc = 4)
plt.ylim(100000)
plt.xlim()
plt.show()
```

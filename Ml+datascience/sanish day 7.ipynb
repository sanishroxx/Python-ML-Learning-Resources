```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axis
fig, ax = plt.subplots()

# Set the axis limits
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

# Create a circle to represent the eclipse
circle = plt.Circle((0, 0), radius=1, color='r',alpha=0.5 )

# Add the circle to the axis
ax.add_artist(circle)

# Generate some random points inside the eclipse
x_inside = np.random.uniform(-1, 1, size=50)
y_inside = np.random.uniform(-1, 1, size=50)

# Generate some random points outside the eclipse
x_outside = np.random.uniform(-2, 2, size=50)
y_outside = np.random.uniform(-2, 2, size=50)

# Plot the points inside the eclipse
ax.plot(x_inside, y_inside, 'o', color='g')

# Plot the points outside the eclipse
ax.plot(x_outside, y_outside, 'o', color='g')

# Display the plot
plt.show()
```

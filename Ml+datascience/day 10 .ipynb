```python
F='Food'; B='Body'; A='Age'
ytrue=[A,B,F,F,A,B,F,F,B,F,B]     
ypred=[A,F,F,B,F,B,F,F,A,B,A]    
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np
import seaborn as sn
```

```python
df=confusion_matrix(ytrue,ypred)
```

```python
df
```

```python
cm=pd.DataFrame(df,columns=np.unique(ytrue),index=np.unique(ypred))
cm.index.name='Act(Row)'
cm.columns.name='Pred(Col)'
```

```python
cm
```

```python
import matplotlib.pyplot as plt
 
plt.figure(figsize=(5,5))
sn.set(font_scale=1.4)
 
sn.heatmap(cm,cmap='Blues',annot=True,annot_kws={"size":16})
```
